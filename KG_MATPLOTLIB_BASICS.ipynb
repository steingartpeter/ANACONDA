```python
#import neccessary libs:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read the gas data 
gd = pd.read_csv('DATA/gas_prices.csv')
```

```python
gd.head(5)
```

```python
# A simple line plot, just assign something to x, and y and => done
plt.plot(gd.Year,gd.Japan)
# Lets add 2 countries, to compare
plt.plot(gd.Year,gd['USA'])
# Or three
plt.plot(gd.Year,gd['Germany'])
```

```python
# Now lets resize the plot
plt.figure(figsize=(9,6))
# Lets say it is too ugly, and pimp it up a bit
plt.plot(gd.Year,gd['Japan'],'bx-')
plt.plot(gd.Year,gd['USA'],'r.-')
plt.plot(gd.Year,gd['Germany'],'g^-')
# Add some lables/ticks
plt.xticks(gd.Year,rotation=35)
plt.xlabel('YEAR')
plt.ylabel('Gall./$')

# As a first step add a title
plt.title("GAS Prices in US $")
# Now add a legend a well
plt.legend(['JPN','USA','GER'])
```
