## Changeing data types

```python
temp = pd.DataFrame({"A":["1","2","3"],"B":[11,12,13],"C":["2020-05-01","2020-06-01","2020-07-01"]})
```

```python
temp.dtypes
temp
```

```python
#convert element "C" to datetime
temp["C"] = pd.to_datetime(temp["C"])
temp.dtypes
```

```python
data010 = uberDf.groupby(['START','STOP']).agg(['mean','min','max']).head(20)
```

```python
#uberDf.dtypes
#uberDf['START_DATE'] = pd.to_datetime(uberDf['START_DATE'])
#uberDf['END_DATE'] = pd.to_datetime(uberDf['END_DATE'])
#uberDf.groupby(['START'])['START_DATE'].agg(['min','max','mean'])
# =======================================================================================================
# get the earliest, and most recent trip date, and the average distance, in miles for each
# starting city: =>
# The main point is, that, for this we need to add a dictionay as an argment for the "agg" function.
# In this dictionary there is the key + value pairs, as:
# COLUMN_NAME : [array, of the functionames pplied to that column]
# =======================================================================================================
data_002 = uberDf.groupby(['START']).agg({'START_DATE':['min','max'],'MILES':['mean']})
#uberDf.tail()
data_002.head(10)
```
